```python
import numpy as np
data = np.loadtxt(fname='../inflammation-01.csv',delimiter=",")
```

```python
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
image = plt.imshow(data)
plt.show()
```

```python
ave_inflammation = np.mean(data, axis=0)
ave_plot = plt.plot(ave_inflammation)
plt.xlabel('time (days)')
plt.ylabel('inflammation score')
plt.title('patient data')
plt.show()
```

```python
max_plot = plt.plot(np.max(data, axis=0))
plt.show()
```

```python
fig = plt.figure(figsize=(10.0, 3.0))
#created a figure that holds multiple plots

fig.suptitle('Title')

sbplt1 = fig.add_subplot(1, 3, 1)
sbplt2 = fig.add_subplot(1, 3, 2)
sbplt3 = fig.add_subplot(1, 3, 3)
#created 3 place holder objects, defining where in this figure object we want to store each sub plot

sbplt1.set_ylabel('average')
sbplt1.plot(np.mean(data, axis=0))

sbplt2.set_xlabel('time (days)')
sbplt2.set_ylabel('max')
sbplt2.plot(np.max(data, axis=0))

sbplt3.set_ylabel('min')
sbplt3.plot(np.min(data, axis=0))

fig.tight_layout
#makes the graph fit tightly next to each other

plt.show()
```

```python
#get all even numbers from 1 to 100
evens = []
#This blank thing is to store it
for ii in range(1,101):
    if (ii % 2 == 0):
        #This is saying if I divid by 2 and the remainder is 0 then...
        evens.append(ii)
        
print(evens)

#get all even numbers from 1 to 100
odds = []
for ii in range(1,101):
    if (ii % 2 == 1):
        odds.append(ii)
        
print("\n")
print(odds)
```
