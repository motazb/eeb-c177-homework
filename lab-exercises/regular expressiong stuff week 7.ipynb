```python
import re
```

```python
def check_string(string):
    charRe = re.compile(r'[^a-zA-Z0-9.\s]')
    #this will find anything that isn't what's bw those brackets
    string = charRe.search(string)
    return not bool(string)
```

```python
check_string("ABCDEFabcdef 133")
```

```python
check_string("%&XD")
```

```python
check_string("&$XD.")
```
