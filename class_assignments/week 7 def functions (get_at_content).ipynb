```python
#Sig figs will be = 2 default unless specified
def get_at_content(dna, sig_figs=2):
    length = len(dna)
    a_count = dna.count('A')
    t_count = dna.count('T')
    at_content = (a_count + t_count) / length
    return round(at_content, sig_figs)

get_at_content("ATGCGCGATCGATCGAATCCG", 5)
```
