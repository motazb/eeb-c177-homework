```python
print(ice_cream['Sarah'])
print(ice_cream['Joe'])
```

```python
food_list = 'pasta pizza bread bread pizza coffee tacos taco tacos burgers burgers burger burgers tacos pasta pasta cheese'.split()
#.split() turns the output to a list instead of a string
```

```python
print(food_list)
```

```python
food_counts = defaultdict(int)
print(food_counts)
```

```python
for food in food_list:
    print(food)
    #This would just print the foods
```

```python
#If we want to count those values we need to specify our index or our values in those foods:
for food in food_list:
    #food_counts[food] = food_counts[food] + 1 this is right but it's long
    food_counts[food] += 1 #the bracket food makes food the key
```

```python
print(food_counts)
```

```python
#If I want the food_counts for just pizza I can just do
print(food_counts['pizza'])
```

```python
city_list = [('TX','Austin'), ('TX','Houston'), ('NY','Albany'), ('NY', 'Syracuse'), ('NY', 'Buffalo'), ('NY', 'Rochester'), ('TX', 'Dallas'), ('CA','Sacramento'), ('CA', 'Palo Alto'), ('GA', 'Atlanta')]
print(city_list)
```

```python
cities_by_state = defaultdict(list)
print(cities_by_state)
#Inside the parenthesis we identify the data structure which is a list (because it has square breackets)
```

```python
#for state, city in city_list:
    #cities_by_state[state] = city_list.append(city)
    #If we do this, city will be appended to city_list. But we want it appended to cities_by_state
    #[state] will be an empty list for that state.
    
for state, city in city_list:
    cities_by_state[state].append(city)
print(cities_by_state)
```

```python
print(cities_by_state.items())
#this will give me the dictionary items, which u can iterate through a loop (like print it out or do a calculation with it)
```

```python
#need a for loop that will print cities and states out
for state, cities in cities_by_state.items():
    #state is the key, cities is the value
    print(state)
    #this will print out all the states
    #If I wanted to print out each cities joined by something like a comma:
    #print(state, ', '.join(cities))
```

```python
for state, cities in cities_by_state.items():
    print(state, ', '.join(cities))
```
