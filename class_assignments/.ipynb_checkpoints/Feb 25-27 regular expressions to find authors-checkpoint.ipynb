```python
with open(inputpath, 'r') as fr:
    data = csv.DictReader(fr, delimiter = '\t')
    header = data.fieldnames
    tax_auth = []
    for row in data:
        tax_auth.append(row['Taxon Author'])
        
#auth_regex = re.compile(r'\(?[a-zA-Z]+-?[a-zA-Z]+')
    #The plus says u can get this one or more times.
    #this regex doesn't recognize year or if name has two '-'
auth_regex = re.compile(r'\(?([\w,\&\-\s]*),\s(\d{4})\)?')
    #re.compile(author, whitespace, 4 digits)
    #\(? - outside of capture group, code will match it, but not capture
    #this regex creates match group 0 and match group 1
```

```python
auth_match = re.match(auth_regex, tax_auth[0])

mystring = 'Tadauchi, Miyanag & Dawut, 2005'
mystring2 = '(Linnaeus, 1758)'
mystring3 = 'Meade-Waldo, 1914'

re.match(auth_regex, mystring).group()
auth_match
```

```python
mystring = 'Tadauchi, Miyanag & Dawut, 2005'
mystring2 = '(Linnaeus, 1758)'
mystring3 = 'Meade-Waldo, 1914'
mystring4 = 'Michener, 1986'

test_strings = [mystring, mystring2, mystring3, mystring4]
```

```python
re.findall(auth_regex, str(test_strings))
    #findall returns all matches
    #search returns only first match
```

```python
for string in test_strings:
    print(re.match(auth_regex, string).group(1))
```

```python
#write a function for regex match
#take each record and return list with two items
#item 1: capture group 1
#item 2: capture group 2

def match_author(record):
    author_match = re.match(auth_regex, record)
    authors = author_match.group(1)
    year = author_match.group(2)
    authors = re.split(', | \&', authors)
    #splits a string given separate
    return[authors, year]
```

```python
match_author(mystring)
```

```python
from collections import defaultdict

author_dict = defaultdict(int)
year_dict = defaultdict(int)

for record in test_strings:
    match_author(record)
    year_dict[record[1]] += 1
```
