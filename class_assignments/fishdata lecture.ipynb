```python
#Answer on website
with open('fishdata.csv', "r") as ff:
    reader = csv.DictReader(ff)
    header = reader.fieldnames #extract the header
    labrid_lengths = []
    with open ('fishdata.csv' + "test" + ".csv", "w") as fw:
        writer = csv.DictWriter(fw, fieldnames = header, delimiter = ",")
        for row in reader:
            if row ["family"] == "Labridae":
                length = row["length"]
                if length != "NA":
                    labrid_lengths.append((float(length)))
                mm = sum(labrid_lengths) / len(labrid_lengths)
                #print(length)
                writer.writerow(row)
        print("The mean length of the labrids is { :.O2f}".format(mm))
```
