```python
import operator
```

```python
#1.
#ascending
d = {1: 2, 3: 4, 5: 6, 6: 7, 0: 0}
#defines dictionary
sorted_d = sorted(d.items(), key=operator.itemgetter(1))
#defines 'sorted_d' to mean sort the dictionary containing items d inside empty parenthesis then fetch the items inside in the operand to be listed inside the parenthesis
print(sorted_d)
#prints sorted_d
```

```python
#descending
sorted_d = dict( sorted(d.items(), key=operator.itemgetter(1),reverse=True))
print(sorted_d)
#this defines sorted_d to mean sort the dictionary containing items d inside empty parenthsis then fetch the items inside in the operand to be listed inside the parenthsis then set the reverse value to true to reverse them.
```

```python
#2.
d = {0:10, 1:20}
#creates dictionary
d.update({2: 30})
#update is a command to add another value to the dictionary. So we're saying add {2: 30} to d
print(d)
```

```python
#3.
dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50, 6:60}
#creates the three dictionaries
```

```python
#4.
dic4 = {}
#dic4 creates a new empty dictionary. 
for d in (dic1, dic2, dic3): dic4.update(d)
#takes dic1, dic2, dic3 and makes a set out of them (d). Then adds d into the empty dic4
print(dic4)
```

```python
def Presence(x):
    #defines entering Presence to check for x.
    if x in dic4:
        print('Present')
        #If x is in dic4, it'll respond with 'Present'
    else:
        print('Not present')
        #If x is not in dic4, it'll respond with 'Not present'
Presence(1)
Presence(1.5)
```

```python
#6.
n=int(input("1nput a number "))
#n is an integer that I input through an "Input a number" slot
d = dict()
#d is an empty dictionary
for x in range (1,n+1):
#for any x between 1 and n+1 (n is the integer I input)
    d[x]=x*x
    #produce a dictionary that multiplies with itself
print(d)
#print the dictionary
```
